```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({
    "Date":pd.date_range(start="2024-01-13",periods=10,freq="d"),
    "Item":1014,
    "measure_1":np.random.randint(1,10,size=10),
    "measure_2":np.random.random(10).round(2),
    "measure_3":np.random.random(10).round(2),
    "measure_4":np.random.randn(10)
})
```

```python
data
```

```python
#adding some missing values in the dataset
data.loc[[2,9],"Item"]=""
```

```python
data
```

```python
data.loc[[2,7,9],"measure_1"]=np.nan
data.loc[[2,3],"measure_2"]=np.nan
data.loc[2,"measure_3"]=np.nan
data.loc[:6,"measure_4"]=np.nan
```

```python
data
```

```python
data
```

```python
data.dropna() #drop the rows that are having the missing values
```

```python
data.isnull().sum()
```

```python
data.dropna(axis=1) #drop the columns having missing values
```

```python
data.dropna(how="all") #if all the values in the row has missing values then drop that row
```

```python
data.dropna(how="any")
```

```python
data.dropna(thresh=4)  #“thresh=4” means that the rows that have at least 4 non-missing values will be kept. 
#The other ones will be dropped
```

```python
data.dropna(subset=["measure_2","measure_3"])
```

```python
data.dropna(subset=["measure_1","measure_2","measure_3"])
```

```python
data.fillna({"Item":1014,
           "measure_1":0})
```

```python
data.fillna(method="bfill")
```

```python
data.fillna(method="ffill")
```

```python
data.fillna(method="bfill",limit=1) # we can limit the number of missing values replaced with this method. If we set the limit parameter as 1, then a missing value can only be replaced with its next value. The second or third following value will not be used for replacement.
```

```python
data=data.astype({
    "Item":pd.Float64Dtype(),
    "measure_1":pd.Float64Dtype()
})
```
